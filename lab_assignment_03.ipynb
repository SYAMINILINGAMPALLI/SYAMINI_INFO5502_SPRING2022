```python
# You code here (Please add comments in the code):

import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
name = []
dealer = []
mileage = []
rating = []
reviews = []
price = []

for i in range (1,60):
    
    
    website = 'https://www.cars.com/shopping/results/?page='+ str(i) +'&page_size=20&dealer_id=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=20&mileage_max=&sort=best_match_desc&stock_type=cpo&year_max=&year_min=&zip=' 
    
    response = requests.get(website)
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    data = soup.find_all('div', {'class' : 'vehicle-card'})
    
    for i in data:

        name.append(i.find('h2').get_text()) 
        dealer.append(i.find('div', {'class':'dealer-name'}).get_text().strip())
        mileage.append(i.find('div', {'class':'mileage'}).get_text())
        rating.append(i.find('span', {'class':'sds-rating__count'}).get_text())
        reviews.append(i.find('span', {'class':'sds-rating__link'}).get_text())
        price.append(i.find('span', {'class':'primary-price'}).get_text())
df = pd.DataFrame({'Name': name, 'Mileage':mileage, 'Dealer':dealer,
                                'Rating': rating, 'Reviews': reviews, 'Price': price})
df

```
